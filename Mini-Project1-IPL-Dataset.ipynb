## Problem Statement :-

## Problem Description :-
- In order to obtain the count of matches in each and every year , we can use the value_counts() function.

- Here we use that function along with the season_year column since we need to obtain the number of matches in one year.

## Problem Statement :-
- Plot a bar graph to depict the number of matches won by each team in the year 2011

## Problem Description :-
- In order to display a bar graph wherein the count of matches won by each team are displayed.
- First,we need to go the year 2011 for that we use df function
- Later ; in order to get the count of wins we use the value_counts() along with the match_winner column
- We convert it into a dataframe.
- In order to get the names of the teams we use the index.tolist() function in order to append the names into an empty list
- Then ; a bar graph is plotted using the values of teams and lr.
- We get the required bargraph.

```python

import pandas as pd
filePath='Match.csv'
def readCsvData(filePath):
    return pd.read_csv(filePath)
df=readCsvData('Match.csv')
import matplotlib.pyplot as plt
pt=df[df['Season_Year'] == 2011]
lr=pt['match_winner'].value_counts()
pd.DataFrame(lr)
print(lr)
print()
teams=[]
teams=lr.index.tolist()
plt.barh(teams,lr,color="Red")
```
